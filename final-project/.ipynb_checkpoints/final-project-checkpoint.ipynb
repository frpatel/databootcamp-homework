```python
import os
import pandas as pd
import requests
import matplotlib as plt
from datetime import datetime, date, timedelta
```

```python
#https://ourworldindata.org/coronavirus-source-data
filepath = 'https://covid.ourworldindata.org/data/full_data.csv'
covid_df = pd.read_csv(filepath)
covid_df.head()
```

```python
cleandf = covid_df.fillna(0)
cleandf
```

```python
maxdate = max(cleandf['date'])
maxdate
```

```python
filter_current_data = cleandf[(cleandf['date']==maxdate) & (cleandf['location']!='World')]
filter_current_data.sort_values(by=['total_cases'], ascending=False).head(10)
```

```python
#df_new_cases = cleandf[['date','location', 'new_cases']]
df_new_cases = cleandf[cleandf['location']!='World'].groupby('date')
df_new_cases.sum()
```
